<h4>Remove null and empty values</h4>

<h5>Check if the columns contain Nan using .isnull() and check for empty strings using .eq(''), then join the two together using the bitwise OR operator |.
Sum along axis 0 to find columns with missing data, then sum along axis 1 to the index locations for rows with missing data.</h5>

```python
missing_cols, missing_rows = (
    (df2.isnull().sum(x) | df2.eq('').sum(x)) #this will count the nan and empty string values in the frame and sum them(it will return a column and index series with missing column name and index)
    .loc[lambda x: x.gt(0)].index #we will locate the above equation in the frame index
    for x in (0, 1) # we will now loop through each axis(i.e, 0 and 1)
) # it will return a tuple of column and index
```

```python
X = df2.drop('music_genre', axis=1)
y = df2['music_genre'].values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=500, multi_class='multinomial')
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('model', model)])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier  
```

```python
rfc = RandomForestClassifier(n_estimators=1000, max_depth=12, criterion='entropy')
```

```python
pipe_rfc = Pipeline([('scaler', StandardScaler()), ('model', rfc)])
```

```python
pipe_rfc.fit(X_train, y_train)
```

```python
pipe_rfc.score(X_test, y_test)
```
